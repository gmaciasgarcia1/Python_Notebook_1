Requirements:
1. (20%) plot two curves: **2sin(x)** in solid line and **2cos(x)** in dotted line
2. (10%) plot three periods for each function as shown in Q1.jpg
3. (10%) plot legend, axis labels (font size is 15), and grid (color is blue, alpha=0.2)
4. (10%) the range of y-axis is [-4, 4]
5. (10%) color: use red when the function is increasing and use green when it is decreasing.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000)
```

```python
sin=2*np.sin(x)
```

```python
cos=2*np.cos(x)
```

```python
plt.plot(x,sin,'-')
plt.plot(x,cos,'--')
plt.ylim([-4,4])
plt.ylabel('Y',fontsize=15)
plt.xlabel('X',fontsize=15)
plt.grid(color='b',alpha=0.2)
plt.legend(['2sin(x)','2cos(x)'])
```

```python
# X=np.random.rand(1000)*100 creates an array X that contains a sample of 1000 random numbers from a uniform distribution,
# each with a value between 0 and 1, multiplied by 100.
# Y = np.random.normal(50, 10, 1000) creates an array Y that contains a sample of 1000 random numbers from a normal 
# distribution with the mean=50 and standard deviation=10
```

```python
X = np.random.rand(1000) * 100
Y = np.random.normal(50, 10, 1000)
```

```python
print(f'Mean X:{np.mean(X)}, Standard Deviation x:{np.std(X)}')
print(f'Mean Y:{np.mean(Y)}, Standard Deviation Y:{np.std(Y)}')
```

```python
plt.hist(X, bins=20, alpha=.5)
plt.hist(Y, bins=20, alpha=.5)
plt.grid(alpha=0.5)
```
